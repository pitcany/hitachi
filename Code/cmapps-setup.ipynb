```python
x = data.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
data = pd.DataFrame(x_scaled)

# columns 2,3 represent states. the rest are observations.
```

```python
# first two columns represent states, the rest are measurements. All of these are normalized.
data.head()
```

```python
measurements=data[data.columns[2:]] 
model = GaussianHMM(n_components=4, n_iter=1000).fit(measurements)
```

```python
data[data.columns[2:]].head()
```

```python
hidden_states=model.predict(measurements)
```

```python
hidden_states
```

```python
P=model.transmat_
```

```python
mus=model.means_
covars=model.covars_
```

```python
sigmas = np.array(np.sqrt(np.array([np.diag(model.covars_[0]),np.diag(model.covars_[1]),np.diag(model.covars_[2]),np.diag(model.covars_[3])])))
```
